```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path

# Adjust this path if needed
fig_dir = Path("figures")

# List your individual story slice files in the order you want them
slice_files = [
    fig_dir / "story_slice1.png",
    fig_dir / "story_slice2.png",
    fig_dir / "story_slice3.png",
    fig_dir / "story_slice4.png",
    fig_dir / "story_slice5.png",
    fig_dir / "story_slice6.png",
]

images = [mpimg.imread(f) for f in slice_files]

fig, axes = plt.subplots(2, 3, figsize=(12, 8))

for ax, img, fname in zip(axes.flat, images, slice_files):
    ax.imshow(img)
    ax.axis("off")
    # Optional: small title per subplot
    # ax.set_title(fname.stem, fontsize=8)

# Tight layout for less whitespace
plt.tight_layout()

out_path = fig_dir / "story_slices_mosaic.png"
plt.savefig(out_path, dpi=300)
print("Saved:", out_path)
plt.close(fig)
```
